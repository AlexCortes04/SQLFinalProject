```python
import pandas as pd

# Define the file path
file_path = 'C:/Users/Pedro/Dropbox/PC/Documents/GitHub/SQLFinalProject/fuel_comsumption.xlsx'

# List all sheet names, excluding the first sheet
excel_file = pd.ExcelFile(file_path)
sheet_names = excel_file.sheet_names[1:]  # Exclude the first sheet

all_data = pd.DataFrame()

for sheet in sheet_names:
    try:
        if sheet.isnumeric(): 
            sheet_data = pd.read_excel(file_path, sheet_name=sheet, skiprows=7)
            sheet_data['Year'] = sheet 
            all_data = pd.concat([all_data, sheet_data])
    except Exception as e:
        print(f"Could not process sheet {sheet}: {e}")

# Drop columns with all NaN values and unnecessary rows
all_data.dropna(axis=1, how='all', inplace=True)
all_data.dropna(axis=0, thresh=2, inplace=True)
relevant_columns = [
    "Unnamed: 0", "Unnamed: 1", "PASSENGER", "MOTOR-", "BUSES",
    "2-AXLE 4-TIRE", "OR MORE", "COMBINATION", "AND", "OR MORE AND",
    "VEHICLES", "Year"
]
all_data = all_data[relevant_columns]

all_data.columns = [
    "Road Category", "Detail", "Passenger Cars", "Motorcycles", "Buses",
    "Other 2-Axle 4-Tire Vehicles", "Single-Unit 2-Axle 6-Tire or More Trucks",
    "Combination Trucks", "Other 2-Axle", "Single-Unit and Combination Trucks",
    "Total Vehicles", "Year"
]

# Drop rows with NaN in 'Year' and 'Road Category' which are unnecessary headers
all_data = all_data.dropna(subset=['Year', 'Road Category']).reset_index(drop=True)

print("Cleaned Combined Data:")
print(all_data.head(10))
```
